```python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
from tqdm import tqdm
from pprint import pprint
from tqdm import tqdm
import pickle
```

```python
def data_segerate(df):
    n = len(df)
    final_data= []
    for i in range(0,n):
        lst =[]
        new_df= df[i].T.drop_duplicates()
        reset_df=new_df.reset_index()
        reset_df.columns=['index','values']
        lst=[reset_df['index'][1].replace('Asset On Auction ','')]+[_.replace('Auction Date ','').replace('Reserve Price','').replace('Earnest Money Deposit','').replace('INR ','').replace('/-','').replace(' to ','TO').replace(' AM','AM').replace(' PM','PM').replace(',','').replace('Legal Status ','').replace('City ','').replace('View Details','').replace('Participate','') for _ in reset_df['values'].values]
#         print(lst)
        final_data.append(lst)
    dfs = pd.DataFrame(final_data, columns = ['Asset On Auction', 'Date and Time','Reserve Price','Legal Status','Earnest Money Deposit','City','details','participate'])
    return dfs
```

```python
url ='https://www.idfcfirstbank.com/assets-on-auction'
driver = webdriver.Chrome()
driver.get(url)
driver.implicitly_wait(5)
soup = BeautifulSoup(driver.page_source,'lxml')
tables = soup.find_all('table')
df=pd.read_html(str(tables))
print(len(df))
pdf_links = []
pdf_links = soup.find_all('a', href=True)
view_links = []
for link in pdf_links:
    href = link['href']
    if '/pdf/' in href:
        view_links.append(href)
view_links.pop(0)
view_links.pop(0)
print(len(view_links))
# print(len(view_links))
# print(view_links)
link_dfs = pd.DataFrame(view_links)
driver.close()
# print(dfs)
```

```python
# len(df)
```

```python
dfs=data_segerate(df)
dt = dfs['Date and Time'].str.split(' ',n=1,expand=True)
dfs['Date'] =dt[0]
dfs['Time'] = dt[1]
dfs.drop(['Date and Time'],axis=1,inplace=True)
result_df = pd.concat([dfs,link_dfs], axis=1)
result_df.drop(['participate','details'],axis=1,inplace=True)
result_df['Time']=result_df['Time'].replace('TO',' to ',regex=True)
# dfs=dfs.merge(link_dfs)
```
